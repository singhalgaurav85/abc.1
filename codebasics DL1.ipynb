```python
# guys if take log of 0 it will give us error or warning like below 
# so to overcome this error we convert our 0 values with very lessor very which is not zero but almost equal to 0 (eg. 0.000000001)
np.log([0])
```

```python
# log 1 is 0 so to overcome this we similarly we convert our all 1 values equal to 1 (eg. 0.9999999999)
np.log([1])
```

```python
# we use this epsilon for converting them as we want epsilon is nothing but a very small value that we define as below

epsilon = 1e-15
```

```python
y_predicted_new = [max(i,epsilon) for i in y_predicted]
```

```python
y_predicted_new   # as we can see that all zero are replaced with epsilon 
```

```python
# similarly now apply it for converting 1 
y_predicted_new = [min(i-1-epsilon) for i in y_predicted_new]
```

```python
y_predicted_new
```

```python
# now we will convert it in numpy array
y_predicted_new = np.array(y_predicted_new)
# see if we apply log on that now
np.log(y_predicted_new)
```

```python
## see if we not aaply all above then what will happen to our y_predicted 
np.log(y_predicted)
```

```python
## now come back to our topic LOG LoSS or binary crossentropy our formula was :
-np.mean(y_true*np.log(y_predicted_new) + (1 - y_true)*(np.log(1-y_predicted_new)))
```

```python
# we can make a function for doing all this in a single better code 
def log_loss(y_true , y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new) + (1 - y_true)*(np.log(1-y_predicted_new)))

log_loss(y_true , y_predicted)
    
    
    
    
```

```python
# for cost function we are taking our previous code for binary_crossentropy

def log_loss(y_true , y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new) + (1 - y_true)*(np.log(1-y_predicted_new)))

    
```

```python
# we will take sigmoid_numpy because it will give us an array of sigmoid values
def sigmoid_numpy(x):
    return 1 / (1 + np.exp(-x))

#sigmoid_numpy(np.array([10,18,21])
    
sigmoid_numpy(np.array([10,1,0]))
```

```python
def gradient_descent(age,affordability,y_true,epochs):
   # we have to find the value of w1,w2,b for minimum loss 
    w1=w2=1
    b=0
    rate = 0.5
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1*age +w2*affordability + b
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true,y_predicted)
        
        # change in weights and bias pls see the fomulas in diary 
        w1_d = (1/n)*np.dot(np.transpose(age),(y_predicted - y_true))
        w2_d = (1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true))
        b_d = np.mean(y_predicted-y_true)
        
        
        w1 = w1 - rate*w1_d
        w2 = w2 - rate*w2_d
        
        b = b - rate*b_d
        
        print(f" epochs:{i}, w1:{w1}, w2:{w2}, b:{b}, loss:{loss}")
    return w1, w2, b
```

```python
# for this we will run the same code above but with some minro change like these see below:
def gradient_descent(age,affordability,y_true,epochs, loss_threshold):
   # we have to find the value of w1,w2,b for minimum loss 
    w1=w2=1
    b=0
    rate = 0.5
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1*age +w2*affordability + b
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true,y_predicted)
        
        # change in weights and bias pls see the fomulas in diary 
        w1_d = (1/n)*np.dot(np.transpose(age),(y_predicted - y_true))
        w2_d = (1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true))
        b_d = np.mean(y_predicted-y_true)
        
        
        w1 = w1 - rate*w1_d
        w2 = w2 - rate*w2_d
        
        b = b - rate*b_d
        
        if loss<=loss_threshold:
            break
        
        print(f" epochs:{i}, w1:{w1}, w2:{w2}, b:{b}, loss:{loss}")
    return w1, w2, b
```
